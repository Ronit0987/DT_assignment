```python
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
import spacy
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd
import re

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

nlp = spacy.load("en_core_web_sm")


CATEGORIES_UBL = {
    "F&B": ["beverage", "cereal", "milk", "dairy", "nutrition", "ingredients"],
    "Manufacturer": [
        "plant", "production capacity", "manufacturing", "certifications",
        "GMP", "ISO", "dosage forms", "oral suspension", "capsule", "tablet", "sachet",
        "women's health", "gut health", "cognitive health"
    ],
    "Brand": ["product details", "pricing", "consumer trust", "brand story", "quality"],
    "Distributor": ["raw materials", "nutraceuticals", "probiotics", "supply chain"],
    "Not Related": [],
    "Formulation": ["capsule", "tablet", "oral suspension", "dosage forms", "sachet", "liquid", "powder"]
}

# Preprocess text
def preprocess_text(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc if not token.is_stop and token.is_alpha])

# Remove stopwords
def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    words = word_tokenize(text.lower())
    return [word for word in words if word.isalnum() and word not in stop_words]

# Extract links from a webpage
def extract_links(url):
    try:
        response = requests.get(url, timeout=10)
        soup = BeautifulSoup(response.text, "html.parser")
        links = [a['href'] for a in soup.find_all('a', href=True)]
        return links
    except Exception as e:
        print(f"Error extracting links: {e}")
        return []

# Scrape website content
def scrape_website(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")
        text = " ".join([p.get_text() for p in soup.find_all("p")])
        return text
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return ""

# Contextual Scoring Function
def get_context_score(content, keyword, window_size=5):
    # Find all occurrences of the keyword in the content
    keyword_positions = [m.start() for m in re.finditer(keyword, content.lower())]
    context_score = 0

    for position in keyword_positions:

        start = max(0, position - window_size)
        end = position + len(keyword) + window_size
        context = content[start:end]


        if any(context_word in context.lower() for context_word in ["nutritional", "health", "benefit","plant","drink"]):
            context_score += 1

    return context_score


def categorize_content(content):
    scores = defaultdict(int)
    keyword_presence = {keyword: 0 for category in CATEGORIES_UBL.values() for keyword in category}
    keyword_context_scores = {keyword: 0 for category in CATEGORIES_UBL.values() for keyword in category}

    words = remove_stopwords(content)

    for category, keywords in CATEGORIES_UBL.items():
        for keyword in keywords:
            if keyword.lower() in content.lower():
                keyword_presence[keyword] = 1
                context_score = get_context_score(content, keyword)
                keyword_context_scores[keyword] = context_score


                if context_score > 0:
                    scores[category] += 2
                else:
                    scores[category] += 1

    category = max(scores, key=scores.get) if scores else "Not Related"
    return category, keyword_presence, keyword_context_scores

def analyze_website(base_url):
    links = extract_links(base_url)
    about_us_page = None
    products_page = None


    for link in links:
        if "about" in link.lower():
            about_us_page = link if link.startswith("http") else base_url + link
        if "product" in link.lower():
            products_page = link if link.startswith("http") else base_url + link


    data = []

    about_us_content = ""
    products_content = ""

    if about_us_page:
        about_us_content = scrape_website(about_us_page)
    if products_page:
        products_content = scrape_website(products_page)

    if about_us_content or products_content:
        combined_content = about_us_content + " " + products_content
        processed_text = preprocess_text(combined_content)
        category, keyword_presence, keyword_context_scores = categorize_content(combined_content)


        data.append({
            "URL": base_url,
            "Text": combined_content[:200],
            "Processed Text": processed_text[:200],
            "Category": category,
            **keyword_presence,
            **keyword_context_scores
        })

    return data


if __name__ == "__main__":
    websites = [
        "https://www.nestle.com",
        "https://www.drreddys.com",
        "https://www.coca-colacompany.com",
        "https://www.pfizer.com",
        "https://www.pepsico.com",
        "https://www.jnj.com",
        "https://www.generalmills.com",
         "https://www.kelloggs.com"

    ]

    all_data = []
    for website in websites:
        print(f"Processing: {website}")
        data = analyze_website(website)
        all_data.extend(data)

    # Convert to DataFrame and print
    df = pd.DataFrame(all_data)
    df = df.fillna(0)  # Replace NaN values with 0 for missing keywords

    # Show the DataFrame
    print(df)
```
